```python
# Import files
from src.database import *
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Pull data from supabase

```python
# Initialise database
db = ImagineAllTheData()

data = db.supabase.table("article_prototype").select("*").execute()
df = pd.DataFrame(data.data)
```

```python
#print(df.head())
print(f"Number of rows: {df.shape[0]}")

# Print different coutrys
print(f"Got articles from coutries: {df['country'].unique()}")

# Create a histogram of the 'country' column
sns.histplot(df['country'], kde=False, discrete=True)
# Set the y-axis to a logarithmic scale
#plt.yscale('log')
# Display the plot
plt.title("Distribution of Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=90)  # Rotate country labels for readability
plt.show()
```
